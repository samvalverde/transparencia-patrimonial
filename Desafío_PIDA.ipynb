# 1. Fase de entendimiento del negocio
Este proyecto se desarrolla en el marco del Desafío PIDA 2025 (Plataforma de Integridad y Datos Abiertos), impulsado por la Organización de los Estados Americanos (OEA), en conjunto con instituciones públicas costarricenses.  
El propósito central de este desafío es promover el uso de analítica avanzada e inteligencia artificial para fortalecer la transparencia, la rendición de cuentas y la prevención de la corrupción en la gestión pública.

Dentro de este contexto, el presente trabajo se enfoca en la detección de anomalías patrimoniales a partir de la integración de dos fuentes de información públicas de alta relevancia nacional:

1. **Registro Nacional**, que contiene información sobre propiedades inscritas, actos registrales, tipo de bien y localización geográfica.  
2. **Contraloría General de la República**, específicamente en las Declaraciones Juradas de Bienes, que describen el patrimonio declarado por funcionarios públicos.

La integración de estas fuentes permite identificar posibles **incongruencias entre los bienes declarados y los registrados**, lo cual constituye un insumo esencial para la supervisión y control del enriquecimiento ilícito.
## (a) Determinar los objetivos del negocio
### i. Objetivos del negocio
El objetivo general del proyecto es desarrollar un modelo de analítica de datos capaz de identificar patrones anómalos en la relación entre los registros de propiedad y las declaraciones de bienes de funcionarios públicos en Costa Rica.

Los objetivos específicos son los siguientes:

- Priorizar la vigilancia patrimonial institucional a nivel agregado y anónimo, focalizando territorios/instituciones con variaciones atípicas para revisión de control interno.
- Fortalecer la transparencia y trazabilidad: proveer un dashboard entendible con tendencias, picos inusuales y composición por rubros, sin exponer datos personales.
- Eficiencia operativa: reducir el tiempo de detección de “zonas de riesgo” de semanas/meses a horas mediante alertas priorizadas.
- Escalabilidad y sostenibilidad: flujo reproducible que hoy corre con datos sintéticos y mañana con datos reales, con actualizaciones periódicas (mensual/trimestral).
- Proveer insumos trazables y explicables para entes de control, con vistas diferenciadas y notas metodológicas visibles.
- Cumplimiento normativo y ético: controles de acceso, anonimización/pseudonimización cuando aplique, y auditoría de consultas.


### ii. Criterios de éxito (en términos del negocio)

## (b) Determinar los objetivos de minería de datos

### i. Objetivos de minería de datos

Desde la perspectiva técnica, la fase de minería de datos busca traducir los objetivos institucionales en procedimientos analíticos reproducibles y medibles.  
El propósito principal es **diseñar, implementar y evaluar un modelo capaz de detectar anomalías patrimoniales** mediante la integración de fuentes de datos heterogéneas y la aplicación de técnicas avanzadas de análisis.

Los objetivos específicos son los siguientes:

- **Integración y normalización multi-fuente:**  
  Construir un modelo de datos unificado que vincule las declaraciones juradas de bienes de la Contraloría General de la República (CGR), los registros de propiedades del Registro Nacional, los datos de contrataciones públicas (SICOP/MIDEPLAN) y variables contextuales del INEC, permitiendo el análisis por periodo, institución y territorio.

- **Ingeniería de atributos orientada al riesgo:**  
  Generar variables derivadas que permitan caracterizar posibles riesgos patrimoniales, tales como la variación interanual del patrimonio, ratios activos/pasivos, densidad de actos registrales, valor medio de propiedades por funcionario o institución, y relación entre patrimonio declarado y montos adjudicados en contrataciones.

- **Detección de anomalías y asignación de puntajes (scoring):**  
  Aplicar métodos no supervisados o semi-supervisados para detectar observaciones atípicas, complementados con modelos supervisados cuando se disponga de etiquetas o casos confirmados. Cada observación deberá acompañarse de un puntaje o score de riesgo explicable.

- **Explicabilidad mínima viable:**  
  Garantizar que cada alerta emitida esté acompañada de indicadores interpretables —por ejemplo, porcentaje de variación en el patrimonio neto, incremento inusual en el valor de inmuebles o picos en la densidad de actos registrales—, apoyándose en técnicas de interpretación de modelos.

- **Operacionalización del prototipo:**  
  Diseñar un flujo reproducible que genere salidas estructuradas (tablas, archivos CSV) listas para visualización o integración con dashboards institucionales.  
  El código debe permitir alternar entre datos sintéticos y reales mediante una simple modificación en la capa de lectura.

- **Evaluación robusta y estabilidad temporal:**  
  Comprobar la consistencia del desempeño del modelo a lo largo de distintos periodos e instituciones, identificando posibles sesgos y garantizando la estabilidad de los resultados en el tiempo.

### ii. Criterios de éxito (desde la perspectiva de minería de datos)

- **Calidad y completitud de datos:**  
  Menos del 5 % de valores faltantes tras la integración y limpieza. Las variables clave deben mantener coherencia temporal y semántica (por ejemplo, correspondencia entre fechas de acto registral y periodos declarativos).

- **Rendimiento del modelo:**  
  Lograr métricas de desempeño satisfactorias en los conjuntos de validación, con precision, recall y F1-score superiores a 0.80, y estabilidad de resultados frente a pequeñas variaciones en los datos.

- **Generalización y escalabilidad:**  
  Mantener la eficacia del modelo al incorporar nuevas instituciones, regiones o periodos sin necesidad de retrenamiento completo, asegurando tiempos de ejecución razonables y bajo consumo de recursos.

- **Interpretabilidad y trazabilidad:**  
  Explicar de forma transparente las anomalías detectadas mediante visualizaciones y variables cuantitativas comprensibles.  
  Registrar las versiones del código, parámetros y resultados intermedios para auditoría y verificación posterior.

- **Reproducibilidad:**  
  Permitir la ejecución del flujo completo en entornos controlados (Jupyter, Python ≥ 3.10) con resultados replicables y documentación técnica clara.

- **Robustez frente al ruido y falsos positivos:**  
  Asegurar tolerancia a datos atípicos o inconsistentes, manteniendo la tasa de falsos positivos por debajo del 10 % en pruebas controladas.

- **Alineación con los objetivos institucionales:**  
  Verificar que las alertas generadas respondan efectivamente a los criterios de riesgo definidos por los entes de control, aportando información útil y accionable para la toma de decisiones.

# 4. Fase de modelado:
Esta fase describe las etapas del entrenamiento y evaluación. Se debe entrenar
cuatro modelos (algoritmos), puede ser de regresión o clasificación.
Debe constar los diferentes experimentos con sus respectivos parámetros con el fin de seleccionar
el mejor modelo con búsqueda de hyperparámetros, (puede usar herramientas específicas para ello
como GridSearchCV, Optuna o MLFlow) . Para ello, no olvide separar los datos en training y testing para
la evaluzación de los mismos, y aplicar CrossValidation como técnica para la selección tomando en
cuenta que el modelo escogido tenga una relación adecuada entre Bias y Variance.
Además, se debe constatar que los modelos no están realizando un sobreajuste.

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt

# Renombrar columnas clave para consistencia
df_cgr_agg = df_cgr_agg.rename(columns={"anio_declaracion": "anio"})
df_registro_agg = df_registro_agg.copy()

# Merge por año y provincia (sin institución)
df_merged = pd.merge(
    df_cgr_agg,
    df_registro_agg,
    on=["anio", "provincia", "canton", "distrito"],
    how="inner"
)

print("Filas resultantes tras merge:", len(df_merged))
display(df_merged.head())

# Ratios de consistencia patrimonial
df_merged["ratio_bienes_vs_patrimonio"] = df_merged["valor_total_bienes"] / df_merged["patrimonio_neto"]
df_merged["ratio_bienes_vs_activos"]     = df_merged["valor_total_bienes"] / df_merged["activos_totales"]
df_merged["densidad_actos_por_func"]     = df_merged["cant_actos_registrales"] / df_merged["cant_funcionarios"]

# Limpieza de infinitos o nulos
df_merged = df_merged.replace([np.inf, -np.inf], np.nan).fillna(0)

# Variaciones por provincia
df_merged = df_merged.sort_values(["provincia", "anio"])

for col in ["valor_total_bienes", "patrimonio_neto", "activos_totales", "pasivos_totales"]:
    df_merged[f"var_{col}"] = df_merged.groupby("provincia")[col].pct_change()

# Diferencia entre variaciones (patrimonio vs bienes)
df_merged["dif_var_bienes_patrimonio"] = df_merged["var_valor_total_bienes"] - df_merged["var_patrimonio_neto"]


# Selección de features para el modelo
features = [
    "ratio_bienes_vs_patrimonio",
    "ratio_bienes_vs_activos",
    "densidad_actos_por_func",
    "var_valor_total_bienes",
    "var_patrimonio_neto",
    "dif_var_bienes_patrimonio"
]

X = df_merged[features].fillna(0)
print("DATA FRAME X")
display(X)
X_scaled = MinMaxScaler().fit_transform(X)
print("DATA FRAME X_SCALED")
display(X_scaled)

# Modelo de anomalías
iso = IsolationForest(
    n_estimators=200, 
    contamination=0.05, 
    random_state=42
    )
df_merged["score_anomalia"] = iso.fit_predict(X_scaled)
df_merged["anomaly_score"] = iso.decision_function(X_scaled)

# Ranking de anomalías por distrito
ranking_distrito = (
    df_merged.groupby("distrito")["anomaly_score"]
    .mean()
    .sort_values()
    .reset_index()
)
ranking_distrito["ranking"] = np.arange(1, len(ranking_distrito) + 1)
print("=== Ranking de distritos más atípicos ===")
display(ranking_distrito.head(10))

plt.figure(figsize=(10, 5))
plt.bar(ranking_distrito["distrito"].head(10), ranking_distrito["anomaly_score"].head(10), color="skyblue")
plt.title("Score promedio de anomalía por distrito")
plt.xlabel("Distrito")
plt.ylabel("Score de anomalía (más bajo = más atípico)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Ranking de anomalías por provincia
ranking = (
    df_merged.groupby("provincia")["anomaly_score"]
    .mean()
    .sort_values()
    .reset_index()
)
ranking["ranking"] = np.arange(1, len(ranking) + 1)

print("\n=== Ranking de provincias más atípicas ===")
display(ranking.head(10))


plt.figure(figsize=(10, 5))
plt.bar(ranking["provincia"], ranking["anomaly_score"], color="salmon")
plt.title("Score promedio de anomalía por provincia")
plt.xlabel("Provincia")
plt.ylabel("Score de anomalía (más bajo = más atípico)")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# Guardar resultados
df_merged.to_csv("analisis_anomalias_territorial.csv", index=False)
ranking.to_csv("ranking_provincias_anomalias.csv", index=False)
```
